scraping  vs crawling?

데이터 스크래핑(data scraping)은 컴퓨터 프로그램이 다른 프로그램으로부터 들어오는 인간이 읽을 수 있는 출력으로부터 데이터를 추출하는 기법

웹 크롤러는 조직적, 자동화된 방법으로 월드 와이드 웹을 탐색하는 컴퓨터 프로그램

페이지 소스 얻어오기 : urllib.request, selenium, requests 등

얻어온 소스를 파싱 : beautifulsoup4

- post 방식 : 홈페이지에 거의 많은 정보가 담겨 있음 (url 바뀌지 x)

- get 방식 : url에 따라 정보가 바뀜

### [melon top100 데이터를 데이터프레임으로 생성 !]

```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
chrome_loc = 'C:/python_lee/chrome/chromedriver.exe'
```

```python
url_target = 'https://www.melon.com/chart/index.htm'
```

```python
driver1.get(url_target)
```

```python
driver1.page_source
```

```python
melon_html = driver1.page_source
```

```python
soup = BeautifulSoup(melon_html, 'html.parser')
```

```python
soup
```

```python
song_lists = soup.find_all('tr')
```

```python
song_lists
```

```python
s = song_lists[0:]
```

```python
for i in range(1, 101):
        ans = s[i].select('span > a')
        song = ans[0].text
        print(song)
```

```python
for i in range(1, 101):
    ans2 = s[i].select('span > a')
    singer = ans2[1].text
    print(singer)
```

```python
import pandas as pd
top100 = pd.DataFrame(columns=['artist', 'song'])
```

```python
for i in range(1, 101):
        ans = s[i].select('span > a')
        ans2 = s[i].select('span > a')
        song = ans[0].text
        singer = ans2[1].text
        top100 = top100.append(pd.DataFrame([[singer, song]], columns=['artist', 'song']), ignore_index=True)
```
