페이지 수 계산

```python
soup.select('.total')[0].text
```

```python
p_num = re.compile('[0-9]+')
```

```python
review_num = int(p_num.findall(soup.select('.total')[0].text)[0]+p_num.findall(soup.select('.total')[0].text)[1])
```

```python
import math
```

```python
# 소수 첫째 자리에서 올림으로 페이지 정보 계산
page_num = math.ceil(review_num/10)
```

```python
page_num
```

2. 모든 페이지 파싱하여 필요 데이터 추출

```python
p_remove = re.compile('\r|\t|\n')
```

```python
soup.select('.score_reple')[0].find(id = '_filtered_ment_0').text
```

```python
p_remove.sub("", soup.select('.score_reple')[0].find(id = '_filtered_ment_0').text)
```

```python
reviews = []
p_remove = re.compile(r'\n|\r|\t')
    
for page in range(1, page_num+1):
    url_page = main_url+str(page)
    res1 = req.urlopen(url_page)
    soup1 = BeautifulSoup(res1, 'html.parser')
    
    i = 0

    while True:
        try:
            a = soup1.select('.score_reple')[i].find(id = ('_filtered_ment_'+str(i))).text
            a2 = p_remove.sub("",a)
            
            reviews.append(a2)
            
            i = i+1
        except IndexError as e:
            print('review {} page done:'.format(page))
            break
```

```python
ids = []
dts = []
star_score = []
sym_num = []
p_num = re.compile('[0-9]+')
    
for page in range(1, page_num+1):
    url_page = main_url+str(page)
    res1 = req.urlopen(url_page)
    soup1 = BeautifulSoup(res1, 'html.parser')
    
    j = 0

    while True:
        try:
            ids.append(soup1.select('dt span')[j].text)
                    
            j = j+1
        except IndexError as e:
            print('id {} page done'.format(page))
            break
    k = 0

    while True:
        try:
            dts.append(soup1.findAll('dt')[k].select('em')[1].text)
                    
            k = k+1
        except IndexError as e:
            print('date {} page done'.format(page))
            break
    l = 0
    
    while True:
        try:
            star_score.append(p_num.findall(soup1.select('.star_score')[l].get_text()))
            l = l+1
        except IndexError as e:
            print('score {} page done'.format(page))
            break
            
    m = 0
    
    while True:
        try:
            sym_num.append(p_num.findall(soup1.select('.btn_area')[m].get_text()))
            m = m+1
        except IndexError as e:
            print('sym {} page done'.format(page))
            break
```

```python
col_name = ['reviews']
import pandas as pd
df = pd.DataFrame(reviews, columns=col_name)
```

```python
df["ids"] = ids
df["dates"] = dts
df["star_score"] = star_score
df["sym_num"] = sym_num

df
```
