### [SEQ1], [SEQ2] 소비전류.주파수.진동값에 대한 상관성

#### 1. SEQ1

```python
SEQ1_df = df[['[SEQ1]  F0 Current', '[SEQ1]  F0 Vibration', '[SEQ1]  RisingTime', '[SEQ1]  FallingTime']]
```

```python
data = SEQ1_df.corr()
fig, ax = plt.subplots( figsize=(7,7) )

mask = np.zeros_like(data, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(data, 
            cmap = 'pink', 
            annot = True,   
            mask=mask,     
            linewidths=.5,  
            cbar_kws={"shrink": .5},
            vmin = -1,vmax = 1  
           )  
plt.show()
```

##### 소비전류.주파수.진동값에 대한 상관성에 대해 알아보고자 한다. 
##### 검사 항목 중 noise관련 부분은 실제 정규성을 띄지 않는것으로 보여 제외한다.

##### [SEQ1]  F0 Vibration과 Current의 상관계수 : 0.81 (양의 상관관계)
##### [SEQ1]  F0 Vibration과 RisingTime의 상관계수 : -0.13 (음의 상관관계)
##### [SEQ1]  F0 Vibration과 FallingTime의 상관계수 : 0.026 (양의 상관관계)


```python
import scipy.stats as stats
stats.pearsonr(x = SEQ1_df['[SEQ1]  F0 Vibration'], y = SEQ1_df['[SEQ1]  F0 Current'])
```

```python
stats.pearsonr(x = SEQ1_df['[SEQ1]  F0 Vibration'], y = SEQ1_df['[SEQ1]  RisingTime'])
```

```python
stats.pearsonr(x = SEQ1_df['[SEQ1]  F0 Vibration'], y = SEQ1_df['[SEQ1]  FallingTime'])
```

##### -피어슨 상관계수를 구하는 함수를 이용해 (상관계수, p값)을 확인하여 결론에 적용시킨다.
##### - p값을 확인하지 않으면 절대 안될 것.. p값이 0.01 혹은 0.05보다 작더라도 정확히 상관계수가 높다, 낮다라고 판단하기는 어렵지만
#####  어느정도 범위 내에서 결론을 내릴 수 있을 것이다. -> Vibration & Current / Vibration & FallingTime 

##### - Vibration & RisingTime은 p값이 7.52로 상관계수 결과를 정할 수 없다.

##### 결론 : 
##### 진동값과 소비전류의 상관관계는 0.81로, 매우 강한 상관관계를 갖는다고 할 수 있다.
##### 상관관계는 약한 관계이지만 진동값이 증가하면 소비전류 또한 증가.

#### 2. SEQ2

```python
X = df['[SEQ2]  F0 Frequency']
Y = df['[SEQ2]  F0 Vibration']
```

```python
cov = (np.sum(X*Y) - len(X)*np.mean(X)*np.mean(Y)) / len(X)
cov
```

```python
corr = cov / (np.std(X) * np.std(Y))
corr
```

##### 결론 : 

##### 주파수와 진동값의 상관관계는 0.58로, 어느 정도 상관관계가 있다.
