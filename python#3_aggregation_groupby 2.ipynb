```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'],
                  'key2': ['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2': np.random.randn(5)})
df
```

```python
grouped = df['data1'].groupby(df['key1'])
grouped
```

```python
grouped.mean()
```

```python
grouped.max()
```

```python
# 연도별 매출액과 이익의 평균 확인하기

grouped1 = df[['data1', 'data2']].groupby(df['key1'])
grouped1
```

```python
grouped1.mean()
```

```python
# 연도별, 월별에 대한 매출액 평균 구해보기

means = df['data1'].groupby([df['key1'], df['key2']]).mean()
```

```python
means
```

```python
means.unstack()
```

```python
means1 = df[['data1', 'data2']].groupby([df['key1'], df['key2']]).mean()
means1
```

```python
means1.unstack()
```

### Pivot Tables and Cross-Tabulation

```python
tips = pd.read_csv('data/tips.csv', sep=',')
```

```python
tips
```

```python
## tip 데이터에 tip_pct (tip / total bill) 컬럼 추가
```

```python
tips['tip_pct'] = tips['tip']/tips['total_bill']
```

```python
tips.pivot_table(index=['day', 'smoker'])  # 요일별, 흡연율 별
```

```python
grouped = tips.groupby(['day', 'smoker'])
```

```python
grouped.agg('mean')  

# = grouped.mean()
# = grouped.agg(np.mean)
```

```python
# agg에 여러 메서드 전달하기
```

```python
func = ['mean', 'median', 'count']

# 이외 집계에 사용되는 function
```

```python
grouped.agg(func)
```

```python
# Q) 시간대별(TIME),요일별(DAY), 파티규모(SIZE), 팁 비율(TIP_PCT)를 알고 싶다면?
```

```python
tips.pivot_table(['tip_pct', 'size'], index=['time', 'day'])
```

```python
# Q) 시간대별(TIME), 요일별(DAY), 흡연여부에 따른 파티규모(SIZE), 팁 비율(TIP_PCT)를 알고 싶다면?
```

```python
 tips.pivot_table(['tip_pct', 'size'],
                 index=['time', 'day'],
                 columns='smoker',
                 margins=True,
                 aggfunc='mean')
```

```python
# 2개 메서드
```

```python
tips.pivot_table('tip_pct', index=['time','smoker'], columns='day', aggfunc=[len, np.sum])   
```

#### 교차분석
#### pandas.crosstab : 범주형 - 범주형 관계

- ex) 요일별 남성 방문객과 여성 방문색 수 집계
- 주요 인수 margin = True / False

```python
# 시간대, 요일별 - 흡연여부별 도수 확인

pd.crosstab([tips.time, tips.day], tips.smoker, margins=True) 
```

```python
pd.crosstab([tips.time, tips.day], tips.smoker, margins=False)

# margins => 합계
```
